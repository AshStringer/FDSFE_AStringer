```python
import numpy as pd
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

import missingno as msno
from sklearn import preprocessing
from sklearn.cluster import KMeans
```

```python
url = "https://www1.nyc.gov/html/gbee/downloads/excel/nyc_benchmarking_disclosure_2017_consumption_data.xlsx"
df = pd.read_excel(url, "Information and Metrics")
```

```python
pd.read_excel(url, "Information and Metrics")
```

```python
features = df[["Year Built", "Electricity Use - Grid Purchase (kBtu)"]]
features
```

```python
initial_shape = features.shape
initial_shape
```

```python
features.info()
```

```python
msno.matrix(features)
```

```python
features.dropna(inplace=True)
features.info()
```

```python
features.describe()
```

```python
features = features[features["Year Built"] > 1900]
features = features[features["Electricity Use - Grid Purchase (kBtu)"] > 0]
```

```python
features.shape
```

```python
features.hist(figsize=(10, 5), bins=100)
```

```python
year_col = "Year Built"
elec_col = "Electricity Use - Grid Purchase (kBtu)"

features[elec_col] = np.log10(features[elec_col])
```

```python
features.hist(figsize=(10, 5), bins=100)
```

```python
features.plot(year_col, elec_col, kind="scatter", alpha=0.1)
```

```python
kn = KMeans(n_clusters=7, random_state=1415, verbose=True)
kn.fit(features)
```

```python
kn.n_clusters
```

```python
kn.cluster_centers_
```

```python
year_col = "Year Built"
elec_col = "Electricity Use - Grid Purchase (kBtu)"

features.plot(year_col, elec_col, kind="scatter", alpha=0.1)
for i in range(kn.n_clusters):
  x = kn.cluster_centers_[i, 0]
  y = kn.cluster_centers_[i, 1]
  plt.plot(x, y, "wx")
```

```python
kn.labels_
```

```python
colors = np.array(["r", "g", "b", "y", "k", "c", "m"])

plt.scatter(features[year_col], features[elec_col], color=colors[kn.labels_], alpha=0.1)
for i in range(kn.n_clusters):
  x = kn.cluster_centers_[i, 0]
  y = kn.cluster_centers_[i, 1]
  plt.plot(x, y, "wx")
plt.xlabel("Year Built")
plt.ylabel("Electricity Use [kBtu]")
plt.axis("equal")
```

```python
features_scaled = preprocessing.scale(features.values, axis=0)
```

```python
features_scaled_df = pd.DataFrame(features_scaled, columns=(year_col, elec_col))
```

```python
features_scaled_df.describe()
```

```python
features_scaled_df.hist(figsize=(10, 5), bins=100)
```

```python
kn = KMeans(n_clusters=7, random_state=1415, verbose=True)
kn.fit(features_scaled_df)
```

```python
colors = np.array(["r", "g", "b", "y", "k", "c", "m"])

plt.scatter(features_scaled_df[year_col], features_scaled_df[elec_col], color=colors[kn.labels_], alpha=0.1)
for i in range(kn.n_clusters):
  x = kn.cluster_centers_[i, 0]
  y = kn.cluster_centers_[i, 1]
  plt.plot(x, y, "wx")
plt.xlabel("Year Built")
plt.ylabel("Electricity Use [kBtu]")
```
